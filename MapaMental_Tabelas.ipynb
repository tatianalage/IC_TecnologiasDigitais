```python
import matplotlib.pyplot as plt
import networkx as nx

# Criar o grafo
G = nx.Graph()

# Adicionando o nó principal
G.add_node("Simple Dublin Core \nElements")

# Adicionando os 15 elementos ao nó principal
elements = [
    "Title", "Creator", "Subject", "Description", "Publisher",
    "Contributor", "Date", "Type", "Format", "Identifier",
    "Source", "Language", "Relation", "Coverage", "Rights"
]

for elem in elements:
    G.add_edge("Simple Dublin Core \nElements", elem)

# Desenhando o grafo
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Layout para o mapa mental
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="#F25922", font_size=14, font_color="black", font_weight="bold", edge_color="gray")
plt.show()
```
