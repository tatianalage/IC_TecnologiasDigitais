```python
import pandas as pd
import numpy as np
```

##**Entrevistas**

```python
# Leitura da página Entrev-Dados Gerais do arquivo Queries na base - Entrevistas
entrevistas_dados_gerais = pd.read_excel('Queries na base - Entrevistas.xlsx', sheet_name='Entrev - Dados gerais')
entrevistas_dados_gerais.info()
entrevistas_dados_gerais.head()
```

Vamos combinar as colunas NM_CID_LOCALIDADE, SG_UF_LOCALIDADE e NM_PAIS_LOCALIDADE em uma nova coluna, removê-las e eliminar a colunas vazia para simplificar a visualização.

```python
# Agrupando NM_CID_LOCALIDADE, SG_UF_LOCALIDADE e NM_PAIS_LOCALIDADE
entrevistas_dados_gerais['NM_CID_LOCALIDADE, SG_UF_LOCALIDADE, NM_PAIS_LOCALIDADE'] = (
    entrevistas_dados_gerais['NM_CID_LOCALIDADE'].astype(str) + ': ' +
    entrevistas_dados_gerais['SG_UF_LOCALIDADE'].astype(str) + ': ' +
    entrevistas_dados_gerais['NM_PAIS_LOCALIDADE'].astype(str)
)
# Removendo as colunas originais e a coluna vazia
entrevistas_dados_gerais.drop(['NM_CID_LOCALIDADE', 'SG_UF_LOCALIDADE', 'NM_PAIS_LOCALIDADE', 'Unnamed: 8'], axis=1, inplace=True)

entrevistas_dados_gerais.head()
```

```python
# Leitura da página Entrev - Entrevistadores do arquivo Queries na base - Entrevistas
entrevistadores = pd.read_excel('Queries na base - Entrevistas.xlsx', sheet_name='Entrev - Entrevistadores')
entrevistadores.info()
entrevistadores.head()
```

Existem algumas linhas referentes à mesma entrevista, cada uma contendo um entrevistador. Vamos reorganizar para que cada entrevista corresponda a apenas uma linha, com os entrevistadores separados por ponto e vírgula.

```python
# Agrupando os entrevistadores por 'NM_TIT_ENTREVISTA'
# Para cada grupo, os valores da coluna 'NM_ENTREVISTADOR' são concatenados em uma única string separada por '; '
entrevistadores_agrupados = entrevistadores.groupby('NM_TIT_ENTREVISTA')['NM_ENTREVISTADOR'].apply(lambda x: '; '.join(x)).reset_index()
# Realizando merge dos dados gerais da entrevista com entrevistadores agrupados
entrevistas_dados_gerais = pd.merge(entrevistas_dados_gerais, entrevistadores_agrupados, on='NM_TIT_ENTREVISTA', how='inner')
entrevistas_dados_gerais
```

```python
# Leitura da página Entrev - Modalidade de consulta do arquivo Queries na base - Entrevistas
modalidade = pd.read_excel('Queries na base - Entrevistas.xlsx', sheet_name='Entrev - Modalidade de consulta')
modalidade.tail()
```

Existem algumas linhas referentes à mesma entrevista, cada uma contendo uma condição de acesso. Vamos reorganizar para que cada entrevista corresponda a apenas uma linha, com as condições de acesso separadas por ponto e vírgula.

```python
# Agrupando as condições de acesso por 'NM_TIT_ENTREVISTA'
# Para cada grupo, os valores da coluna 'DS_CONDICAO_ACESSO' são concatenados em uma única string separada por '; '
modalidade_agrupada = modalidade.groupby('NM_TIT_ENTREVISTA')['DS_CONDICAO_ACESSO'].apply(lambda x: '; '.join(x)).reset_index()

# Realizando merge dos dados gerais da entrevista com a modalidade agrupada
entrevistas_dados_gerais = pd.merge(entrevistas_dados_gerais, modalidade_agrupada[['NM_TIT_ENTREVISTA', 'DS_CONDICAO_ACESSO']], on='NM_TIT_ENTREVISTA', how='inner')
entrevistas_dados_gerais
```

```python
# Leitura da página Entrev - Data e duração do arquivo Queries na base - Entrevistas
data_duracao = pd.read_excel('Queries na base - Entrevistas.xlsx', sheet_name='Entrev - Data e duração')
data_duracao.info()
data_duracao.head()
```

```python
# Combinando as colunas QT_HR_SESS_GRAV e QT_MIN_SESS_GRAV acrescentando o formato '%h %min'
data_duracao['QT_HR_SESS_GRAV, QT_MIN_SESS_GRAV'] = data_duracao['QT_HR_SESS_GRAV'].astype(str) + 'h' + data_duracao['QT_MIN_SESS_GRAV'].astype(str) + 'min'
# Convertendo para datetime a data da sessão
data_duracao['DT_SESS_GRAV'] = pd.to_datetime(data_duracao['DT_SESS_GRAV'])
# Definindo o formato %d/%m/%Y
data_duracao['DT_SESS_GRAV'] = data_duracao['DT_SESS_GRAV'].dt.strftime('%d/%m/%Y')
```

```python

data_duracao_agrupada = data_duracao.groupby('NM_TIT_ENTREVISTA').agg({
    'QT_HR_SESS_GRAV, QT_MIN_SESS_GRAV': lambda x: ', '.join(x.astype(str)),
    'DT_SESS_GRAV': lambda x: ', '.join(x.astype(str))
}).reset_index()

# Realizando merge dos dados gerais da entrevista com a data_duração agrupada
entrevistas_dados_gerais = pd.merge(entrevistas_dados_gerais, data_duracao_agrupada[['NM_TIT_ENTREVISTA', 'QT_HR_SESS_GRAV, QT_MIN_SESS_GRAV', 'DT_SESS_GRAV']], on='NM_TIT_ENTREVISTA', how='inner')
entrevistas_dados_gerais
```

```python
# Leitura da página Entrev - Temas do arquivo Queries na base - Entrevistas
entrevistas_temas = pd.read_excel('Queries na base - Entrevistas.xlsx', sheet_name='Entrev -Temas')
entrevistas_temas.head()
```

Como uma mesma entrevista, representada pelo mesmo título, contém várias linhas cada uma com um descritor, vamos reorganizá-las atribuindo apenas uma linha para cada entrevista, separando os descritores por ponto e vírgula.

```python
# Agrupando as ocorrências dos Descritores pelo Título da Entrevista e criando uma única linha para cada entrevista
descritores_agrupados = entrevistas_temas.groupby('NM_TIT_ENTREVISTA')['DS_DEL'].apply(lambda x: '; '.join(x)).reset_index()

# Realizando a mesclagem (merge) dos DataFrames com base em 'NM_TIT_ENTREVISTA'
entrevistas_dados_gerais = entrevistas_dados_gerais.merge(descritores_agrupados[['NM_TIT_ENTREVISTA', 'DS_DEL']], on='NM_TIT_ENTREVISTA', how='left')

entrevistas_dados_gerais
```

Vamos realizar o tratamento dos dados faltantes na coluna 'DS_DEL' para evitar erros na aplicação do filtro e selecionar as entrevistas que possuem os descritores Portugal ou Comunidade dos Países de Língua Portuguesa nos Descritores.

```python
# Tratamento dos dados faltantes na coluna DS_DEL
entrevistas_dados_gerais['DS_DEL'] = entrevistas_dados_gerais['DS_DEL'].fillna('')

# Filtrando as entrevistas que possuem 'Portugal' ou 'Comunidade dos Países de Língua Portuguesa' nos Descritores
temas_portugal = entrevistas_dados_gerais[entrevistas_dados_gerais['DS_DEL'].str.contains('Portugal','Comunidade dos Países de Língua Portuguesa')]
temas_portugal.info()
```

Vamos realizar o tratamento dos dados faltantes na coluna 'DS_RESUMO_FICHA_TEC' para evitar erros na aplicação do filtro e selecionar as entrevistas que contém no Resumo 'Memória das Ciências Sociais em Portugal', 'Cientistas sociais de países de Língua Portuguesa: histórias de vida' e 'Portugal'.

```python
# Tratamento dos dados faltantes na coluna DS_RESUMO_FICHA_TEC
entrevistas_dados_gerais['DS_RESUMO_FICHA_TEC'] = entrevistas_dados_gerais['DS_RESUMO_FICHA_TEC'].fillna('')

# Filtrando as entrevistas que possuem "Memória das Ciências Sociais em Portugal", "Cientistas sociais de países de Língua Portuguesa: histórias de vida" e "Portugal"
resumo_portugal = entrevistas_dados_gerais[entrevistas_dados_gerais['DS_RESUMO_FICHA_TEC'].str.contains('Memória das Ciências Sociais em Portugal|Cientistas sociais de países de Língua Portuguesa: histórias de vida|Portugal')]
resumo_portugal.info()
```

Vamos aplicar o fitro 'Portugal' na coluna combinada NM_CID_LOCALIDADE, SG_UF_LOCALIDADE, NM_PAIS_LOCALIDADE

```python
# Filtrando as entrevistas que possuem 'Portugal' na localidade
localidade_portugal = entrevistas_dados_gerais[entrevistas_dados_gerais['NM_CID_LOCALIDADE, SG_UF_LOCALIDADE, NM_PAIS_LOCALIDADE'].str.contains('Portugal')]
localidade_portugal.info()
```

```python
# Concatenar entrevistas selecionadas por palavras-chave no resumo, localidade e tema
entrevistas_portugal = pd.concat([resumo_portugal, localidade_portugal, temas_portugal])

# Eliminar duplicatas
entrevistas_portugal = entrevistas_portugal.drop_duplicates()

# Ordenar as entrevistas por número de entrevista
entrevistas_portugal = entrevistas_portugal.sort_values(by='CD_ENTREVISTA')

# Mapear os valores da coluna 'IN_TIP_ENTREVISTA'
entrevistas_portugal['IN_TIP_ENTREVISTA'] = entrevistas_portugal['IN_TIP_ENTREVISTA'].map({1: 'História de vida', 0: 'Temática'})

# Adicionar colunas 'NM_TIPO_UNIDADE_DOCUMENTAL' e 'PUBLISHER'
entrevistas_portugal['NM_TIPO_UNIDADE_DOCUMENTAL'] = 'Entrevista'
entrevistas_portugal['PUBLISHER'] = 'FGV CPDOC'

# Definir uma lista com a nova ordem das colunas desejadas
nova_ordem_colunas = [
    'PUBLISHER', 'NM_TIT_ENTREVISTA', 'NM_ENTREVISTADOR', 'DT_SESS_GRAV','QT_HR_SESS_GRAV',
    'QT_MIN_SESS_GRAV', 'DS_CONDICAO_ACESSO', 'DS_DEL', 'DS_RESUMO_FICHA_TEC','CD_ENTREVISTA',
    'URL', 'NM_TIT_PROJETO', 'NM_TIPO_UNIDADE_DOCUMENTAL', 'IN_TIP_ENTREVISTA','NM_CID_LOCALIDADE, SG_UF_LOCALIDADE, NM_PAIS_LOCALIDADE'
]

# Reorganizar as colunas do DataFrame conforme a nova ordem
entrevistas_portugal = entrevistas_portugal.reindex(columns=nova_ordem_colunas)

# Ajustando o índice do DataFrame para começar em 1 em vez de 0
entrevistas_portugal.index = np.arange(1, len(entrevistas_portugal) + 1)

# Exibir o DataFrame resultante
entrevistas_portugal
```

```python
# Exportar arquivo para Excel
entrevistas_portugal.to_excel('Entrevistas_Portugal.xlsx')
```

## **Audiovisual**

```python
# Leitura da página 'Geral Audiovisual' na planilha 'Queries na base - Audiovisual & Manuscritos.xlsx'
audiovisual_geral = pd.read_excel('Queries na base - Audiovisual & Manuscritos.xlsx', sheet_name = 'Geral Audiovisual')
audiovisual_geral.info()
```

```python
# Filtrando os audiovisuais que possuem Portugal ou Comunidade dos Países de Língua Portuguesa nos Descritores
temas_audiovisual = audiovisual_geral[audiovisual_geral['DESCRITORES'].str.contains('Portugal|Comunidade dos Países de Língua Portuguesa', na=False)]

# Filtrando os audiovisuais que possuem Portugal ou Comunidade dos Países de Língua Portuguesa no Título
titulo_audiovisual = audiovisual_geral[audiovisual_geral['DS_TITULO_AVI'].str.contains('Portugal|Comunidade dos Países de Língua Portuguesa', na=False)]

# Filtrando os audiovisuais que possuem Portugal na Localidade
localidade_audiovisual = audiovisual_geral[audiovisual_geral['NM_LOCALIDADE'].str.contains('Portugal', na=False)]

# Concatenando os DataFrames filtrados e removendo duplicatas
portugal_audiovisual = pd.concat([temas_audiovisual, titulo_audiovisual, localidade_audiovisual]).drop_duplicates()

# Removendo colunas desnecessárias ('Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13') do DataFrame
portugal_audiovisual = portugal_audiovisual.drop(columns=['Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13'])
```

```python
# Criando uma nova coluna 'PUBLISHER' e definindo todos os seus valores como 'FGV CPDOC'
portugal_audiovisual['PUBLISHER'] = 'FGV CPDOC'

# Criando uma nova coluna NM_TIPO_UNIDADE_DOCUMENTAL e definindo todos os seus valores como 'Audiovisual'
portugal_audiovisual['NM_TIPO_UNIDADE_DOCUMENTAL'] = 'Audiovisual'

# Definindo uma nova ordem para as colunas do DataFrame
nova_ordem_colunas = [ 'PUBLISHER','DS_TITULO_AVI', 'DS_PERIODO_PRODUCAO_AVI', 'NR_DOCUMENTOS_AVI',
                       'DS_FISICA_AVI', 'AUDIOVISUAL', 'NM_FUN', 'NM_TIPO_UNIDADE_DOCUMENTAL', 'DS_SER',
                       'NM_LOCALIDADE']

# Reorganizando as colunas do DataFrame de acordo com a nova ordem definida
portugal_audiovisual = portugal_audiovisual.reindex(columns = nova_ordem_colunas)

# Ajustando o índice do DataFrame para começar em 1 em vez de 0
portugal_audiovisual.index = np.arange(1, len(portugal_audiovisual) + 1)

# Exibir o Dataframe resultante
portugal_audiovisual
```

```python
# Exportar arquivo para Excel
portugal_audiovisual.to_excel('Audivisual_Portugal.xlsx')
```

## **Manuscritos**

```python
# Leitura da página Geral Manuscrito do arquivo Queries na base - Audiovisual & Manuscritos
manuscrito_geral = pd.read_excel('Queries na base - Audiovisual & Manuscritos.xlsx', sheet_name = 'Geral Manuscritos')
manuscrito_geral.info()
```

 Filtro para Descritores e para Resumo com as palavras "Portugal" e "Comunidade dos Países de Língua Portuguesa"

```python
# Filtrando os manuscritos que possuem "Portugal" e "Comunidade dos Países de Língua Portuguesa" nos Descritores
temas_manuscrito = manuscrito_geral[manuscrito_geral['DESCRITORES'].
                                    str.contains('Portugal|Comunidade dos Países de Língua Portuguesa', na=False)]

# Filtrando os manuscritos que possuem "Portugal" e "Comunidade dos Países de Língua Portuguesa" no Resumo
resumo_manuscrito = manuscrito_geral[manuscrito_geral['DS_RESUMO_MAN'].
                                     str.contains('Portugal|Comunidade dos Países de Língua Portuguesa', na=False)]

# Concatenar os manuscritos com os temas e resumo selecionados
portugal_manuscrito = pd.concat([temas_manuscrito, resumo_manuscrito])
```

```python
# Tratamento para dados ausentes
portugal_manuscrito['NR_DOCUMENTOS_TIPO_MAN'].fillna(0, inplace=True)

# Acrescentar f. no formato
portugal_manuscrito['NR_DOCUMENTOS_TIPO_MAN'] = portugal_manuscrito['NR_DOCUMENTOS_TIPO_MAN'].astype(int).astype(str) + 'f.'

# Acrescentar "doc(s)" no formato
portugal_manuscrito['NR_DOCUMENTOS_MAN'].astype(str) + ' doc(s)'

# Criando uma nova coluna PUBLISHER e definindo todos os valores como 'FGV CPODC
portugal_manuscrito["PUBLISHER"] = 'FGV CPDOC'

# Criando uma nova coluna NM_TIPO_UNIDADE_DOCUMENTAL
portugal_manuscrito['NM_TIPO_UNIDADE_DOCUMENTAL'] = 'Textual'

# Definindo uma nova ordem para as colunas do DataFrame
colunas_manuscrito = ['PUBLISHER', 'DS_PERIODO_PRODUCAO_MAN', 'NR_DOCUMENTOS_MAN',
           'NR_DOCUMENTOS_TIPO_MAN', 'DESCRITORES', 'DS_RESUMO_MAN', 'MANUSCRITO',
           'NM_FUN', 'DS_SER', 'NM_TIPO_UNIDADE_DOCUMENTAL']

# Reorganizando as colunas do DataFrame de acordo com a nova ordem definida
portugal_manuscrito = portugal_manuscrito.reindex(columns=colunas_manuscrito)

# Ajustando o índice do DataFrame para começar em 1 em vez de 0
portugal_manuscrito.index = np.arange(1, len(portugal_manuscrito) + 1)

# Exibir o Dataframe resultante
portugal_manuscrito
```

```python
# Exportar para Excel
portugal_manuscrito.to_excel("Textual_Portugal.xlsx")
```
